```python
import numpy as np
import jax.numpy as jnp
import pylab as plt
```

```python
# rng = np.random.default_rng(17) # the most random number
rng = np.random.default_rng(42) # the second most random number
```

```python
# make some latent parameters and the data therefrom
D = 2
sigma_data = 0.45
true_z = jnp.array(rng.normal(size=D)) # magic
true_epsilon = sigma_data * jnp.array(rng.normal(size=D))
y_data = true_z + true_epsilon
print("true r is", jnp.sqrt(true_z @ true_z))
```

```python
# make a likelihood function; need to vectorize this
def log_like(theta, r, yy = y_data, ss = sigma_data):
    chi = (yy - r * jnp.array([jnp.cos(theta), jnp.sin(theta)])) / ss
    return -0.5 * chi @ chi
```

```python
# plot the log likelihood as a function of angle for various r
nn = 600
thetas = np.arange(0. + 0.5 / nn, 4. * np.pi, 1. / nn)
rs = np.arange(0.8, 1.21, 0.1)[::-1]
mll = -np.inf
for r in rs:
    lls = np.array([log_like(t, r) for t in thetas])
    if np.max(lls) > mll:
        mll = np.max(lls)
    plt.plot(thetas, lls, label=f"{r:0.2f}")
plt.xlim(0., 4. * np.pi)
plt.legend()
```
